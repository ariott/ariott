```python
from youtube_transcript_api import YouTubeTranscriptApi

def extract_subtitles(video_id):
    try:
        transcript_list = YouTubeTranscriptApi.get_transcript(video_id)
        subtitles = ""
        for transcript in transcript_list:
            subtitles += transcript['text'] + " "
        return subtitles.strip()
    except Exception as e:
        print("Error:", e)

def extract_subtitles_with_timestamps(video_id):
    try:
        transcript_list = YouTubeTranscriptApi.get_transcript(video_id)
        subtitles = ""
        for transcript in transcript_list:
            start = transcript['start']
            duration = transcript['duration']
            text = transcript['text']
            subtitles += f"{start} --> {start + duration}\n{text}\n\n"
        return subtitles.strip()
    except Exception as e:
        print("Error:", e)

if __name__ == "__main__":
    video_id = input("Enter YouTube video ID: ")
    subtitles = extract_subtitles(video_id)
    subtitles_with_timestamps = extract_subtitles_with_timestamps(video_id)

    print("\nSubtitles:")
    print(subtitles)

    print("\nSubtitles with timestamps:")
    print(subtitles_with_timestamps)
```

```python
import difflib

def find_corresponding_words(word1, text1, word2, text2):
    """
    Find corresponding words in two texts based on similarity.
    """
    words_text1 = text1.split()
    words_text2 = text2.split()

    matcher1 = difflib.SequenceMatcher(None, words_text1, [word1])
    matcher2 = difflib.SequenceMatcher(None, words_text2, [word2])

    # Get matching blocks
    matching_blocks1 = matcher1.get_matching_blocks()
    matching_blocks2 = matcher2.get_matching_blocks()

    corresponding_words1 = [words_text1[start1:start1+length1] for start1, _, length1 in matching_blocks1 if length1 > 0]
    corresponding_words2 = [words_text2[start2:start2+length2] for start2, _, length2 in matching_blocks2 if length2 > 0]

    return corresponding_words1, corresponding_words2
```

```python
def main():
    # Example texts
    text1 = "The quick brown fox jumps over the lazy dog."
    text2 = "A quick brown fox jumps over a lazy dog."

    # Take user input for two words
    word1 = input("Enter a word from the first text: ")
    word2 = input("Enter a word from the second text: ")

    # Find corresponding words
    corresponding_words1, corresponding_words2 = find_corresponding_words(word1, text1, word2, text2)

    # Print the corresponding words in both texts
    if corresponding_words1 and corresponding_words2:
        print(f"Corresponding words for '{word1}' in the first text: {corresponding_words1}")
        print(f"Corresponding words for '{word2}' in the second text: {corresponding_words2}")
    else:
        print("No corresponding words found.")
```
